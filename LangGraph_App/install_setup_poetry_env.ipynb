### Set up poetry and virtual environment
  + install conda environment with python 12.4
    `conda create -n py3124 python=3.12.4`
  + activate the virtual environment
    `conda activate py3124`
  + create poetry environment using conda environment
    `poetry env use (Get-Command python).Source`
  + poetry will create a virtual environment under the current folder. open the pyproject.toml file and change the python requirement to ">=3.12.4,<4.0", as shown below:
    + 
  + install python-dotenv, black isort, langchain, langchain-openai, langgraph by                                        
    `poetry add python-dotenv black isort langchain langchain-openai langgraph`
